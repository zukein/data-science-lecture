## Mean Shiftの仕組み
---
特徴空間で、指定した距離内の密度が最大化するようにクラスタ中心を決める手法。
1. 初期のクラスタ中心を選ぶ
 - 全てのサンプルをクラスタ中心に選ぶ
 - 格子状に領域を分割し、一定以上のサンプル数が含まれる区間をクラスタ中心に選ぶ
1. 指定された距離内のデータ点を元に密度が最大になる方向 (指定範囲内のサンプル重心 = 平均) にクラスタ中心を移動する
1. 収束するまで 2 を繰り返す
1. 指定された距離内にクラスタ中心が複数ある場合には密度の高いものを残す
